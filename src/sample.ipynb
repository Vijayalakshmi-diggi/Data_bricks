```python
a=[1,2,3,3,4]
b=[]
dup=[]
for i in a:
    if i not in b:
        b.append(i)
    else:
        dup.append(i)
print(b)
print(dup)
```

```python
def reverse(s):
    if len(s)==0:
        return s
    return reverse(s[1:])+s[0]
print(reverse("python"))
```
