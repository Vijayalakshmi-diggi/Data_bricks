# list methods (append(),extend(),remove(),pop(),insert(),index(),sort(),reverse(),clear(),count(),copy())

```python
my_list=[1,2,'viji',5,4]
print(my_list)
print(my_list[2])

my_list.append(10)
print(my_list)

my_list.insert(4,'geetha')
print(my_list)

my_list.pop(1)
print(my_list)

my_list.remove(5)
print(my_list)

my_list[2]='vivek'
print(my_list)

print(my_list[:3])
print(my_list[3:])
print(my_list[::2])
print(my_list[1:5:3])

my_list.extend([11,12,9])
print(my_list)

print(my_list.index('viji'))

lst=my_list.copy()
print(lst)

my_list.clear()
print(my_list)
```

```python
numbers=[1,4,2,2,4,3,5]

print(numbers.count(2))

numbers.sort()
print(numbers)

numbers.reverse()
print(numbers)

numbers.append([1,2])
print(numbers)
```

tuple methods (count(),index()) reason tuple is immutable howevr index can be used with start and end argument for searching the index, slicing can be done in tuple


```python
t=(1,2,5,4,3,2)
print(type(t))
print(t)

print(t.count(2))
print(t.index(3))
print(t.index(2,2))

print(t[:3])
print(t[1:3])
print(t[::2])
print(t[1:5:3])
```

set methods (17) commonly used ,add(),remove(),discard()

```python
s={1,3,2,6,4,5,5,8,}
print(s)

s.remove(6)
print(s)

s.add(7)
print(s)

s.update([10,11,12])
print(s)

s.discard(4)
print(s)

s.pop()
print(s)

s1={3,4,6,0,1}
print(s.union(s1))

print(s.intersection(s1))

print(s.difference(s1))

print(s.symmetric_difference(s1))

print(s.isdisjoint(s1))

print(s.issubset(s1))

print(s.issuperset(s1))

s.clear()
print(s)
```
