```sql
use catalog my_sql;
use schema joins;
CREATE TABLE employees (
    id INT,
    name STRING,
    salary INT
)
USING DELTA;

INSERT INTO employees
VALUES
(1,'Alice',50000),
(2,'Bob',60000);

DESCRIBE DETAIL employees;

SELECT *
FROM employees VERSION AS OF 0;

CREATE TABLE customer (
    id INT,
    name STRING,
    city STRING
)
USING DELTA;

INSERT INTO customer
VALUES
(1, 'Alice', 'Delhi'),
(2, 'Bob', 'Mumbai'),
(3, 'John', 'Chennai');

select *from customer;

CREATE TABLE customer_updates (
    id INT,
    name STRING,
    city STRING
)
USING DELTA;

INSERT INTO customer_updates
VALUES
(2, 'Bob', 'bangalore'),
(5, 'vivek', 'Hyderabad');

delete from customer_updates where city='bangalore'
and id=2;

select *from customer_updates;

merge into customer as target
using customer_updates as source
on target.id=source.id 

when matched then 
update set*

when not matched then 
insert (id,name,city)
values(source.id,source.name,source.city);

select *from customer;
```
